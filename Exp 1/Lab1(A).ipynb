# Aim: To perform the basic operations on the Fuzzy sets
## Theory: 
Fuzzy Set: A fuzzy set allows elements to have degrees of membership between 0 and 1, representing uncertainty or partial membership.

Key Operations on Fuzzy Sets:

Union: The maximum membership value of each element in two sets: mu_A ∪ B(x) = max(mu_A(x), mu_B(x))

Intersection: The minimum membership value of each element: mu_A ∩ B(x) = min(mu_A(x), mu_B(x))

Complement: Reverses membership values: mu_¬A(x) = 1 - mu_A(x)

Sum: Adds membership values and subtracts their product: mu_A + B(x) = mu_A(x) + mu_B(x) - mu_A(x) * mu_B(x)

Difference: Minimum of the membership in A and the complement of B: mu_A - B(x) = min(mu_A(x), 1 - mu_B(x))

Vector Product: Element-wise product of membership values: mu_A * B(x) = mu_A(x) * mu_B(x)

```python
class Fuzzyset:
  def __init__(self):
    self.data = {}
    self.get_elements()

  def get_elements(self):
    self.n = int(input("Enter the number of ordered pair: "))
    print("Enter elements for the set\n")
    for _ in range(self.n):
      key = int(input("Enter the key: "))
      value = float(input("Enter the value: "))
      self.data[key] = value

  def union(self, other):
    res = {key: max(self.data.get(key, 0), other.data.get(key, 0)) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def intersection(self, other):
    res = {key: min(self.data.get(key, 0), other.data.get(key, 0)) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def complement(self):
    return {key: 1 - value for key, value in self.data.items()}

  def sum(self, other):
    res = {key: (self.data.get(key, 0) + other.data.get(key, 0) - (self.data.get(key, 0) * other.data.get(key, 0))) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def vector_product(self, other):
    res = {key: self.data.get(key, 0) * other.data.get(key, 0) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def difference(self, other):
    res = {key: min(self.data.get(key, 0), 1 - other.data.get(key, 0)) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def bounded_sum(self, other):
    res = {key: min(1, self.data.get(key, 0) + other.data.get(key, 0)) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def bounded_difference(self, other):
    res = {key: max(0, self.data.get(key, 0) + other.data.get(key, 0) - 1) for key in set(self.data.keys()) | set(other.data.keys())}
    return res

  def cross_product(self, other):
    res = [[min(self.data.get(i, 0), other.data.get(j, 0)) for j in other.data] for i in self.data]
    return res
```

```python
def main():
  print("For Set A: ")
  A = Fuzzyset()
  print("\nFor Set B: ")
  B = Fuzzyset()

  operations = {
      1: lambda self: self.intersection(B),
      2: lambda self: self.union(B),
      3: lambda self: self.complement(),
      4: lambda self: self.sum(B),
      5: lambda self: self.difference(B),
      6: lambda self: self.cross_product(B),
      7: lambda self: self.vector_product(B),
      8: lambda self: self.bounded_sum(B),
      9: lambda self: self.bounded_difference(B),
  }

  print('''
  1. For intersection
  2. For Union
  3. For complement
  4. For sum
  5. For difference
  6. For cross_product
  7. For vector product
  8. For bounded_sum
  9. For bounded_difference
  ''')

  while True:
    try:
      choice = int(input("Enter the choice: "))
      if choice in operations:
        print(operations[choice](B))
      else:
        print("Enter correct choice")
    except ValueError:
      print("Invalid input. Please enter an integer.")

    c = input("Do you want to continue(Y/N): ")
    if c.lower() not in ['y', 'yes']:
      break

if __name__ == "__main__":
  main()
```
