## Union

```python
def union(dic1,dic2):
  res={}
  keys=set()
  keys.update(list((dic1.keys())))
  keys.update(list(dic2.keys()))
  for key in keys:
    try:
      res[key]=max((dic1[key]),dic2[key])
    except:
      pass
  return res
```

## Intersection

```python
def intersection(dic1,dic2):
  res={}
  keys=set()
  keys.update(list((dic1.keys())))
  keys.update(list(dic2.keys()))
  for key in keys:
    try:
      res[key]=min((dic1[key]),dic2[key])
    except:
      pass
      #print("Element does not exist in both dictionary")
  return res
```

## Complement

```python
def complement(dic):
  res=dic.copy()
  for key,value in dic.items():
    res[key]=1-value
  return res
```

#Vector Product
`Ua(x)*Ub(x)`

```python
def vector_product(dic1,dic2):
  res={}
  keys=set()
  keys.update(list((dic1.keys())))
  keys.update(list(dic2.keys()))
  for key in keys:
    try:
      res[key]=(dic1[key]*dic2[key])
    except:
      pass
      #print("Element does not exist in both dictionary")
  return res
```

# Difference

```python
def difference(dic1,dic2):
  return intersection(dic1,complement(dic2))
```

# Bounded Sum

```python
def bounded_sum(dic1,dic2):
  res={}
  keys=set()
  keys.update(list((dic1.keys())))
  keys.update(list(dic2.keys()))
  for key in keys:
    try:
      res[key]=min(1,(dic1[key]+dic2[key]))
    except:
      pass
      #print("Element does not exist in both dictionary")
  return res
```

#Bounded Difference

```python
def bounded_difference(dic1,dic2):
  res={}
  keys=set()
  keys.update(list((dic1.keys())))
  keys.update(list(dic2.keys()))
  for key in keys:
    try:
      res[key]=max(0,(dic1[key]+dic2[key]-1))
    except:
      pass
      #print("Element does not exist in both dictionary")
  return res
```

# Cross product

```python
def cross_product(dic1,dic2):
  keys1=list(dic1.keys())
  keys2=list(dic2.keys())
  res=[]
  for i in keys1:
    temp=[]
    for j in keys2:
      temp.append(min(dic1[i],dic2[j]))
    res.append(temp)
```
