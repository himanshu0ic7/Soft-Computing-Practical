## Aim: To implement Linear Regression with one variable and hence predicting the house of a price have 1200 sq.ft

## Theory:

In this experiment, we perform linear regression using a single feature to predict housing prices based on the size of the house. Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables.

The following components are covered in this experiment:
1. **Dataset Preparation:**
   - The input data (`x_train`) represents the size of houses in 1000 square feet.
   - The target data (`y_train`) represents the corresponding prices of the houses in thousands of dollars.

2. **Model Representation:**
   - The hypothesis function is defined as `f_wb = w * x + b`, where `w` is the weight (slope), and `b` is the bias (intercept).
   - Initial values for `w` and `b` are set to 200 and 100, respectively.

3. **Visualizing the Data:**
   - The data points are plotted to show the relationship between house size and price.
   - A line representing the model's predictions is plotted along with the actual data points.

4. **Prediction:**
   - The model is used to predict the price of a house with a size of 1200 sq.ft.
   - The predicted price is calculated using the hypothesis function, yielding an estimated value of $340,000.


```python
import numpy as np
import matplotlib.pyplot as plt
x_train = np.array([1.0, 2.0]) #x_train is the input variable (size in 1000 square feet)
y_train = np.array([300.0, 500.0]) #y_train is the target (price in 1000s of dollars)
print(f'x_train = {x_train}')
print(f'y_train = {y_train}')
```

```python
# m is the number of training examples
m = len(x_train)
print(f'Number of training examples is: {m}')
i = 0 # Change this to 1 to see (x^1, y^1)
x_i = x_train[i]
y_i = y_train[i]
print(f'(x({i}), y({i})) = ({x_i}, {y_i})')
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x',c='r')
# Set the title
plt.title('Housing Prices')
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 200
b = 100
def compute_model_output(x, w, b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b
  return f_wb
tmp_f_wb = compute_model_output(x_train, w, b,)
```

```python
# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Predictions')
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
# Set the title
plt.title('Housing Prices')
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10, 6))
plt.plot(x_train, tmp_f_wb, c='b', label='Our Predictions')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
data = np.array(list(zip(x_train, y_train)))
table = plt.table(cellText=data, colLabels=['Size (1000 sqft)', 'Price (in 1000s of dollars)']
                  ,cellLoc='center', loc='bottom', bbox=[0.1, -0.3, 0.8, 0.2])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.legend()
plt.show()
```
